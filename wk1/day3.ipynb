```python
def repeat(num_times):
    def decorator_repeat(func):
        def wrapper(args):
            for _ in range(num_times):
                result = func(args)
            return result
        return wrapper
    return decorator_repeat

```

```python
@repeat(num_times=10)
def myprint(name):
    print(f"바보,{name}")
```

```python
myprint("알ㄹ야")
```

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(10)
```

```python
def fibo(n):
    if n==1:
        return 1
    if n==2:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
```

```python
%%time
fibo(30)
```

```python
# 제너레이터, co-routine
def fib():
    a,b=0,1
    while True:
        yield a # routine 짜는.
        a,b = b, a+b
```

```python
%%time
fib_gen = fib()
first_10 = [next(fib_gen)for _ in range(40)]
```
