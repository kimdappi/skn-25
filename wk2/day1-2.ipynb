### 큐 생성

```python
from collections import deque
dq =deque()
dq.append(10)
dq.appendleft(20)
print("1.",dq)
dq.append(30)
print("2.",dq)
dq.pop()
print("3.",dq)
dq.popleft()
print("4.",dq)
```
