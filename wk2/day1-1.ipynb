```python

class Student: # 안에서 돌아가는 로직이라는 뜻 def__init__
    def __init__(self, name, korean, math, english, science):
        self.name = name 
        self.korean = korean
        self.math = math
        self.english = english 
        self.science = science 

    def get_data(self):
        return self.name
```

```python
class Student:
    def __init__(self,name):
        self._name=name

    @property
    def math(self):
        return self.name
    
    @math.setter
    def math(self,value):
        print("setter")
        if 0<=value <=100:
            self._name= value
        else:
            raise ValueError("retry")
    
    @math.getter
    def name(self):
        print("getter")
        return self.name

    def __str__(self): #print 출력을 위한 내장함수
        return self._name
    
b = Student('엔코아')
```
