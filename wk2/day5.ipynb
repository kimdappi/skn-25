## 호텔 정보 넣기

```python
sql= """CREATE TABLE hotel_ratings (
   rating VARCHAR(100) NOT NULL,  -- 등급 인정처
   hotel_type VARCHAR(50) NOT NULL,        -- 호텔업 구분
   final_rating VARCHAR(20) NOT NULL,      -- 결정 등급
   rating_date DATE NOT NULL,     -- 등급 결정일
   region VARCHAR(50) NOT NULL,            -- 지역
   hotel_name VARCHAR(100) NOT NULL,       -- 호텔명
   room_count INT NOT NULL,                -- 객실수
   address VARCHAR(200) NOT NULL,          -- 주소
   phone_number VARCHAR(20),               -- 전화번호
   website_url VARCHAR(255),                -- 홈페이지
   primary key (hotel_name, address)
);
"""

cursor.execute(sql) 
```

```python
import csv
file = "./문화체육관광부_전국호텔현황_20230405.csv"
sql = "INSERT INTO hotel_ratings VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
with open(file, 'r', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    header = next(csv_reader)
    # print(f"헤더: {header}")
    for row_index, row_data in enumerate(csv_reader):
        if len(row_data) == 10:
            try:
                cursor.execute(sql, row_data)
            except Exception as e:
                print(e)
```
