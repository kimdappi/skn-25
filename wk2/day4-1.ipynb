```python
import pandas as pd
```

```python
kbo_df=pd.read_csv("./kbo.csv")
```

```python
kbo_df.info()
```

```python
kbo_df['생년월일'].isnull().sum()
kbo_df['생년월일'].apply(lambda x : len(x) !=13).sum()
```

```python
# kbo_df['생년월일']=pd.to_datetime(kbo_df['생년월일'],format='%Y년 %m월 %d일')
# kbo_df['days']=datetime.now()-kbo_df['생년월일']
```

```python
from datetime import datetime


kbo_df['days'] = kbo_df['생년월일'].apply(lambda x :  \
            (datetime.now() - datetime.strptime(x, "%Y년 %m월 %d일")).days)
kbo_df.sort_values(by=['Team','days'],ascending=[True,False])
```

```python
kbo_df['Team'].unique()
```

```python
kbo_df[kbo_df['Team']=='고양 히어로즈']
kbo_df.loc[kbo_df['Team']=='고양 히어로즈','Team'] = '키움 히어로즈'
kbo_df['Team'].unique()
```

```python
kbo_df[(kbo_df['Team']=='LG 트윈스') & (kbo_df['포지션']=='투수(우투우타)')]
kbo_df[(kbo_df['Team']=='LG 트윈스') | (kbo_df['포지션']=='투수(우투우타)')]
```

```python
kbo_df.to_csv('./kbo2.csv',index=False)
```

```python
import re
p=re.compile("[가-힣a-zA-Z]+초")
kbo_df['출신초등학교'] = kbo_df['경력'].apply(lambda x : p.findall(x.split('-')[0])[0] if len(p.findall(x.split('-')[0])) > 0 else '-')
```

```python
kbo_df.to_csv('./kbo2.csv',index=False)
```
